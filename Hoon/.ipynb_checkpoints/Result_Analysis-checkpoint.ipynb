# Class Distribution on Submission Result 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def check_distribution(file_path):
    submission = pd.read_csv(file_path)
    plt.figure(figsize=(10, 5))
    ax = sns.countplot(x=submission['target'])
    plt.title('Distribution of class')
    plt.xlabel('Class')
    plt.ylabel('Count')
    
    # 막대 위에 수치값 표기
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 3, f'{height}', ha='center', size=9)
    plt.show()
```

```python
check_distribution('./submissions/bert_klue_sub_070583.csv')
```
