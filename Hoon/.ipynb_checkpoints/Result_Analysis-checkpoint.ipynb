```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def check_distribution(file_path):
    submission = pd.read_csv(file_path)
    plt.figure(figsize=(10, 5))
    sns.countplot(x=submission['target'])
    plt.title('Distribution of class')
    plt.xlabel('Class')
    plt.ylabel('Count')
```
