```python
# train 데이터의 최대 길이를 구함
train_len = [len(x.split()) for x in train_df['conversation']]
MAX_LEN = max(train_len)
```

```python
MAX_LEN
```

```python
result_df = pd.DataFrame()
```

```python
sub_class = '불안'
file_list = glob.glob(fr"**/*TS_{sub_class}*/**/*.tsv", recursive=True)
len(file_list)
```

```python
for i, file_name in enumerate(file_list):
    print(i)
    df = pd.read_csv(file_name, sep='\t')
    sample = '\n'.join(df['utterance_text'])
    if len(sample.split()) <= MAX_LEN:
        continue
    result_df.loc[len(result_df), ['class', 'conversation']] = [f'일반대화_{sub_class}', sample]
```
